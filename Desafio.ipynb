```python
import pandas as pd
import matplotlib.pyplot as plt
from openpyxl import load_workbook
```

```python
# Criando os dados
data = {
    'Produto': [f'P {chr(65 + i)}' for i in range(20)],
    'Valor': [100 + i * 5 for i in range(20)],
    'Quantidade Vendida (Mensal)': [150 - i * 5 for i in range(20)]
}
```

```python
# Criando um DataFrame com os dados
df = pd.DataFrame(data)
```

```python
# Criando um gráfico de barras com o valor médio de cada produto
plt.figure(figsize=(10, 6))
plt.bar(df['Produto'], df['Valor'])
plt.xlabel('Produto')
plt.ylabel('Valor')
plt.title('Valor médio por produto')
plt.xticks(rotation='horizontal')
```

```python
plt.savefig('grafico.png')
```

```python
df['Valor total de vendas'] = df['Valor'] * df['Quantidade Vendida (Mensal)']
```

```python
df.to_excel('perfumes.xlsx', index=False)
```
